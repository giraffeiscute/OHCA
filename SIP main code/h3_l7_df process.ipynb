```python
import pandas as pd
import numpy as np
import os
import h3 
import re
```

```python
# building_df = pd.read_csv('OpenStreetMap/B_features.csv')
poi_df_B = pd.read_csv('OpenStreetMap/B_features.csv')

poi_df_B = poi_df_B[['osmid','building', 'geometry','amenity']]

# 初始化列表
lat = []
lon = []
h3_l7 = []

# 遍歷每一行
for i in range(poi_df_B.shape[0]):
    # 提取經緯度
    coordinates = re.findall("\d+\.?\d*", poi_df_B.geometry[i])
    if len(coordinates) < 1:
        continue  # 如果座標數據不完整，跳過該行
    else:
        # 提取緯度和經度
        latitude = float(coordinates[1])  # 緯度
        longitude = -float(coordinates[0])  # 經度
        lat.append(latitude)
        lon.append(longitude)
        
        # 轉換為 H3 L7 id
        h3_l7.append(h3.geo_to_h3(latitude, longitude, resolution=7))


# 新增列到有效的 DataFrame
poi_df_B['lat'] = lat
poi_df_B['lon'] = lon
poi_df_B['h3_l7'] = h3_l7

# 刪除 'geometry' 列
poi_df_B = poi_df_B.drop('geometry', axis=1)
poi_df_B = poi_df_B[poi_df_B.amenity.isna()]
poi_df_B = poi_df_B[poi_df_B.building != 'yes']
poi_df_B = poi_df_B.drop(['amenity'], axis=1) 
poi_df_B
```

```python
# building_df = pd.read_csv('OpenStreetMap/B_features.csv')
poi_df_A = pd.read_csv('OpenStreetMap/A_features.csv')
# poi_df = poi_df[['osmid', 'amenity', 'ele', 'denomination', 'brand', 'cuisine',
#                     'healthcare', 'parking', 'geometry', 'height']
poi_df_A = poi_df_A[['osmid', 'amenity', 'geometry']]

# 初始化列表
lat = []
lon = []
h3_l7 = []

# 只挑選有效的行
valid_indices = []  # 用來記錄有效行的索引
# 遍歷每一行
for i in range(poi_df_A.shape[0]):
    # 提取經緯度
    coordinates = re.findall("\d+\.?\d*", poi_df_A.geometry[i])
    if len(coordinates) < 1:
        continue  # 如果座標數據不完整，跳過該行
    else:
        # 提取緯度和經度
        latitude = float(coordinates[1])  # 緯度
        longitude = -float(coordinates[0])  # 經度
        lat.append(latitude)
        lon.append(longitude)
        
        # 轉換為 H3 L7 id
        h3_l7.append(h3.geo_to_h3(latitude, longitude, resolution=7))
        
        # 記錄有效行的索引
        valid_indices.append(i)



# 新增列到有效的 DataFrame
poi_df_A['lat'] = lat
poi_df_A['lon'] = lon
poi_df_A['h3_l7'] = h3_l7

# 刪除 'geometry' 列
poi_df_A = poi_df_A.drop('geometry', axis=1)
poi_df_A['building'] = poi_df_A['amenity'] 
poi_df_A = poi_df_A.drop(['amenity'], axis=1) 
poi_df_A
```

```python
unique_valuesB = poi_df_B['building'].unique()
unique_valuesA = poi_df_A['building'].unique()
# 找出 A 和 B 中的重複值
duplicate_values = set(unique_valuesA) & set(unique_valuesB)  # 使用集合交集操作

# 將結果轉成列表或其他格式
duplicate_values_list = list(duplicate_values)
duplicate_values_list
```

```python
poi_df = pd.concat((poi_df_A, poi_df_B))

min_lat = min(poi_df['lat'])
max_lat = max(poi_df['lat'])
min_lon = min(poi_df['lon'])
max_lon = max(poi_df['lon'])
poi_df
```

```python
ohca_df = pd.read_csv('OHCAs.csv')
h3_l7 = []

for i in range(ohca_df.shape[0]):
    h3_l7.append(h3.geo_to_h3(ohca_df.Latitude[i], ohca_df.Longitude[i], resolution=7))

ohca_df['h3_l7'] = h3_l7

if min_lat < min(ohca_df['Latitude']): min_lat = min(ohca_df['Latitude'])
if max_lat > max(ohca_df['Latitude']): max_lat = max(ohca_df['Latitude'])
if min_lon < min(ohca_df['Longitude']): min_lon = min(ohca_df['Longitude'])
if max_lon > max(ohca_df['Longitude']): max_lon = max(ohca_df['Longitude'])

ohca_df = ohca_df.drop_duplicates(subset=['ReceivedTime', 'Latitude', 'Longitude'])
ohca_df['ReceivedTime'] = pd.to_datetime(ohca_df['ReceivedTime'])
ohca_df['ReceivedTime'] = ohca_df['ReceivedTime'].apply(lambda x: x.date())
ohca_df
```

```python
ohca_df.to_csv('ohca_df.csv' , index=False)
```

```python
cols = np.concatenate((
    poi_df.building.unique(),  # 從 poi_df 中提取唯一的建築類型
))
len(cols)  # 計算列數

h3_l7_df = pd.DataFrame(data={'id': np.unique(np.concatenate((poi_df.h3_l7.unique(), ohca_df.h3_l7.unique())))})

h3_l7_df[cols] = 0

for i in range(poi_df.shape[0]):  # 遍歷 poi_df 中的每一行
    h3_l7_id = poi_df.iloc[i]['h3_l7']  # 獲取當前行的 h3_l7 值
    h3_l7_df.loc[h3_l7_df['id'] == h3_l7_id, poi_df.iloc[i]['building']] += 1  # 根據 building 類型累加計數
```
