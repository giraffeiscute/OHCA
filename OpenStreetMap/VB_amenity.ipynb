```python
import networkx as nx
import osmnx as ox
# import folium 
import math
%matplotlib inline
ox.__version__
import pandas as pd
import numpy as np
# pd.set_option('display.max_columns', 1000)
import re
```

```python
df = pd.read_csv('A_features.csv')
```

```python
lat = []
lon = []
for i in range(df.shape[0]):
    lat.append(float(re.findall("\d+\.?\d*", df.geometry[i])[1]))
    lon.append(float(re.findall("\d+\.?\d*", df.geometry[i])[0]))

df['lat'] = lat
df['lon'] = lon
```

```python
min_lat = min(df['lat'])  # 最小緯度
max_lat = max(df['lat'])  # 最大緯度
min_lon = min(df['lon'])  # 最小經度
max_lon = max(df['lon'])
```
